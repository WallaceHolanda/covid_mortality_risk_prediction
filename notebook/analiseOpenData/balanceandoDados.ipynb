```python
def get_key(my_dict, val):
    for key, value in my_dict.items():
         if val == value:
             return key
    return -1
```

```python
def formatarMapaBalanceado(mapa, coluna, base, chave_1, chave_2, proporcao):
    balanceamentoMapa = {}
    balanceamentoMapa[str(proporcao)] = len(base)
    
    valores = base[coluna].value_counts()
    
    balanceamentoMapa[mapa[chave_1] + str(proporcao)] = valores[chave_1]
    balanceamentoMapa[mapa[chave_2] + str(proporcao)] = valores[chave_2]
    
    return balanceamentoMapa
```

```python
def juntarMapas(mapa_1, mapa_2, mapa_3):
    mapa_1.update(mapa_2)
    mapa_1.update(mapa_3)
    return pd.DataFrame([mapa_1])
```

```python
# Variáveis de apoio
evolucaoValores = {0: 'Óbito', 1: 'Curado', 2: 'Internado'}
tabelaDeBases = {}
siglaEvolucao = "oc"
```

```python
tabelaBalanceados = pd.DataFrame()
for i in range(0, len(basesObitoCurado)):
    mapaEstado = {}
    mapaEstado['UF'] = re.compile( "sus-(.*?)-pp").search(basesObitoCurado[i]).group(1)
    print("UF ", mapaEstado['UF'])
    base = pd.read_excel("BasesPreProcessadas/" + basesObitoCurado[i])
    
    valor_1 = get_key(evolucaoValores, 'Óbito')
    valor_2 = get_key(evolucaoValores, 'Curado')
    
    base6040 = pp.balanceamento(base, 'evolucaoCaso', valor_1, 40, valor_2)
    mapaBase6040 = formatarMapaBalanceado(evolucaoValores, 'evolucaoCaso', base6040, valor_1, valor_2, '60/40')
    pp.salvarBaseDeDados(base6040, "BasesBalanceadas/"+ siglaEvolucao +"-6040-" + mapaEstado['UF'] + ".xlsx")
    
    base7030 = pp.balanceamento(base, 'evolucaoCaso', valor_1, 30, valor_2)
    mapaBase7030 = formatarMapaBalanceado(evolucaoValores, 'evolucaoCaso', base7030, valor_1, valor_2, '70/30')
    pp.salvarBaseDeDados(base7030, "BasesBalanceadas/"+ siglaEvolucao +"-7030-" + mapaEstado['UF'] + ".xlsx")
    
    tabelaBalanceados = pp.juntarBases([tabelaBalanceados, juntarMapas(mapaEstado, mapaBase6040, mapaBase7030)])
    
tabelaBalanceados
pp.salvarBaseDeDados(tabelaBalanceados, "obito-curado-balanceamento.xlsx")
```
