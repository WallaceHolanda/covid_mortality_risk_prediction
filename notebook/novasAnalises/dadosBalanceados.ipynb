```python
import pandas as pd
import numpy as np
```

```python
dadosSuportePositivo = pd.read_excel("../../database/novosDados/balanceados/suporte_positivo_pp_Jan2021.xlsx")
dadosSuporteNegativo = pd.read_excel("../../database/novosDados/balanceados/suporte_negativo_pp_Jan2021.xlsx")
```

```python
# Imprimindo os dataframes
dadosSuportePositivo.head()
dadosSuporteNegativo.head()
```

```python
# Verificando as dimensões
print("Suporte - Dados Negativos", dadosSuporteNegativo.shape)
print("Suporte - Dados Positivos", dadosSuportePositivo.shape)
```

```python
selecionados = dadosSuportePositivo.sample(144)
selecionados.shape
```

```python
selecionados.head()
```

```python
covidData = pd.merge(dadosSuporteNegativo, selecionados, how = 'outer')
covidData.shape
```

```python
covidData.head()
```

```python
import sklearn
covidData_shuffled = sklearn.utils.shuffle(covidData)
covidData_shuffled.head()
```

```python
covidData_shuffled.reset_index(inplace=True, drop = True)
covidData_shuffled.head()
```

```python
write = pd.ExcelWriter('balanceados_70_30_Jan2021.xlsx')
covidData_shuffled.to_excel(write, 'dados', index = False)
write.save()
```
