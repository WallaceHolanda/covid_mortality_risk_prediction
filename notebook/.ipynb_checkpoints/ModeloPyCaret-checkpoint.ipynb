```python
import pandas as pd
import numpy as np

from pycaret.classification import *
```

```python
covidData = pd.read_excel("../database/dados-preprocessados-novos-organizados.xlsx")
covidData
```

```python
covidData['EVOLUCAO'] = covidData['EVOLUCAO'].map({1:0, 2:1})
print(covidData['EVOLUCAO'])
```

```python
covidData.dtypes
```

```python
setup_classificao = setup(data = covidData, target = 'EVOLUCAO', train_size=0.7)
```

```python
get_config('X')
```

```python
melhores_modelos = compare_models(n_select= 20)
```

```python
melhores_modelos[0]
```

```python
random_forest_tunado = tune_model(melhores_modelos[0])
```

```python
modelo_final = finalize_model(random_forest_tunado)
```
